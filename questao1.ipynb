```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ia_vende_normalmente.csv', sep=';')

df
```

```python
df = df.drop(columns=['cliente_id', 'nome_cliente'])
df.sort_values('data_hora_venda')

df
```

```python
x = df['days_to_next_sale'].plot.kde()
plt.show()
```

```python
df['valor_venda'] = df['valor_venda'].apply(lambda x: x.replace('$','')).apply(lambda x: x.replace(',','.')).apply(lambda x: x.replace('R','')).astype(float)
```

```python
df["valor_venda"] = pd.to_numeric(df["valor_venda"])
plot2 = df['valor_venda'].plot.line()
plt.show()
```

```python
df.groupby(pd.PeriodIndex(df['data_hora_venda'], freq="M"))['valor_venda'].mean().reset_index()
```
